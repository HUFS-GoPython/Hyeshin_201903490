### Frequency Distribution
- 단어 빈도수를 활용할 수 있는 많은 메소드 제공
- 인간 언어 -> 컴퓨터가 인지
    - 단어 수
    - 특정단어의 빈도 수
- 자료형은 딕셔너리

```python
fdist = nltk.FreqDist(book.text1)
fdist
```

```python
fdist.most_common(10)
```

```python
fdist['whale'], book.text1.count('whale')
#둘이 같은 기능
```

```python
fdist.freq('whale')
```

```python
fdist.plot(50, cumulative = True) #False로 두고 그릴 수도 있음
```

```python
# 1번만 나타나는 단어들

len(fdist.hapaxes()) # 중복이 안 돼있음, 대소문자는 구분돼있음
```

### list comprehension
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- list + for_loop 한 줄로 해결하는 방식
- [A for A' in B]
    - B에 있는 A에서 A를 가진 리스트를 반환
    - B: 기존의 리스트
    - A': B의 자료 형태
    - A: 새로운 리스트의 요소

- 조건+ : [A for A' in B if <조건식>]
    - [A for A' in B if <조건식> and <조건식>,...]

```python
[d for d in drinks]
```

```python
# text1 알파벳 수가 10이상인 단어

[word for word in book.text1 if len(word) >= 10]
```

```python
# text4에서 'gnt'를 포함하는 단어 반환

[word for word in book.text4 if 'gnt' in word]
```

```python
# text1에서 'ableness'로 끝나는 단어 반환
[word for word in book.text1 if word.endswith('ableness')]
```

```python
# text3에서 대문자로 시작하는 단어들 중에 길이가 3이상인 단어 -> 중복 없이

set([word for word in book.text3 if word[0].isupper() and len(word) >= 3])
```

```python
# istitle을 쓸 수도 있음

set([word for word in book.text3 if word.istitle() and len(word) >= 3])
```

```python
# text4에서 알파벳인 단어 중, 길이가 4 이상이고 빈도수가 5이상인 단어만
# fdist는 변수 명일 뿐
fdist = nltk.FreqDist(book.text4)
[len(word) for word in book.text4 if word.isalpha() and len(word) >= 4 and fdist[word] >= 5]
```

### 출력 포멧팅
- f'str....{var}' (f,F 둘다 가능)
- %char (옛날 방식)
    - '...%s' %var
- '...{}'.format{var}

```python
# 제일 단순한 방법은
name = 'lee'
age = 18
print('My name is', name)
```

```python
# 2번과 3번은 가독성의 차이임. 넣어주는 게 나을 지 뒤로 뺴는 게 나을지

print(f'My name is {name}, I am {age} years old.')
print('My name is {}, I am {} years old.'.format(name,age))
```

```python
a = 0.555555
print('My name is %.3f' %a)  # 소수점 셋쨰자리까지 
```
